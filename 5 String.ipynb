```python
#String concatenate
print('Shubham'+ 'Kumar')
print('Shubham'+ ' '+'Kumar')
```

```python
#string replicate

print('shubham '* 5)
print( 5*'shubham ')
```

```python
#string length
print(len('shubham kumar'))
```

```python
# slicing
print('shubham kumar'[0])
print('shubham kumar'[1])
print('shubham kumar'[4])
print('shubham kumar'[0:5])
print('shubham kumar'[0:7])
print('shubham kumar'[8:13])
print('shubham kumar'[0:])
print('shubham kumar'[:13])
print('shubham kumar'[:])
print('shubham kumar'[-1]) #r
print('shubham kumar'[-5]) #k
print('shubham kumar'[-5:-1])#kuma
print('shubham kumar'[-5:]) #kumar
```

```python
#case conversion

print("Shubham KUmar".upper())
print("Shubham KUmar".lower())
print("Shubham KUmar".isupper())
print("Shubham KUmar".islower())
print("SHUBHAM KUMAR".isupper())
print("shubham kumar".islower())
```

```python
# string stripping : remove from starting and end
print("  shubham  kumar   ")
print("  shubham  kumar   ".strip())
print("  sh  ubham  kumar   ".strip())
print("    my  name   is   shubham  kumar     ".strip())
```

```python
# string replace

print("shubham  kumar".replace( 's', 'S'))
print("shubham  kumar".replace( ' ', '-'))
print(("shubham  kumar".replace( 's', 'S')).replace('k', 'K'))
```

```python
# string count
print(("shubham  kumar".count('h')))
print(("shubham  kumar".count('sh')))
print(("shubham  kumar".lower().count('h')))
print(("Ashish jangra".count('a')))
print(("Ashish jangra".lower().count('a')))
print(("Ashish jangra".count('A')))
print(("Ashish jangra".upper().count('A')))
```

```python
# string find: gives index
print("shubham kumar".find('u'))
```

```python
# string check
print("shubham kumar".isalpha()) #false
print("shubham".isalpha())
print("shubham890".isalnum())
print("3424".isnumeric())
print("12354".isdigit())
print("shubham kumar".islower())
print("SHUBHAM".isupper())
```

```python
#string capitalize: make 1st char of sentence to upper
print("shubham kumar".capitalize())
```

```python
# string title
print("shubham kumar".title())
print("my name is shubham kumar".title())
```

```python
# check for start and end
print("shubham kumar".startswith("sh"))
print("shubham kumar".endswith('ar'))
```

```python
#string adjust
print("shubham kumar".center(13, '*')) # center m roght m jaada * hoga
print("shubham kumar".ljust(20, '*'))
print("shubham kumar".rjust(20, '*'))
```
