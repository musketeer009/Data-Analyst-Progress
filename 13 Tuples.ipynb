```python
tpl = (1,2,3,4,5,6,7,9)
print(tpl)
```

```python
# accessing
print(tpl[3])
```

```python
# slicing
print(tpl[2:6])
```

```python
# concatenate
tpl1 = (1,2,3,4,5)
tpl2 = (4,5,6,7,8,9,0)
tpl = tpl1 + tpl2
print(tpl)
```

```python
# iterating
for i in tpl:
  print(i)

for i in range(len(tpl)):
  print(tpl[i])
```

```python
# checking the element
print(2 in tpl)
```

```python
# frequency

print(tpl.count(4))
```

```python
# finding index

print(tpl.index(3))
print(tpl.index(4))
print(tpl.index(5))
print(tpl.index(6))
```

```python
# multiply
tpl = (1,2,3,4,5)
print(tpl*2)
```

```python
my_list = list((1, 2, 3, 4, 5))  # Using `list()`
print(my_list)
```
