1. Check if a number is even or odd.

2. Check if a number is positive, negative, or zero.

```python
num = int(input("Enter a number:"))
if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

3. Determine if a character is a vowel or consonant

```python
ch = input("Enter a char:")
ch= ch.lower()
if ch == 'a' or ch == 'e' or ch == 'i' or ch == 'o' or ch == 'u':
    print("Vowel")
else:
    print("Consonant")
```

4. Check if a year is a leap year.

```python
year = int(input("Enter year:"))
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not Leap Year")
```

```python
num = int(input("Enter a number:"))
if num > 0:
    if num % 2 == 0:
        print("Positive Even")
    else:
        print("Positive Odd")
else:
    print("Not Positive")
```

9. Determine if a temperature in Celsius is freezing, cold or warm

```python
temp = int(input("Enter temp:"))
if temp < 0:
    print("Freezing")
elif temp < 15:
    print("Cold")
elif temp < 25:
    print("Warm")
else:
    print("Hot")
```

11. Multi-elif: Determine time of day by hour (0-23)


```python
hour = int(input("Enter time:"))
if hour < 6:
    print("Early Morning")
elif hour < 12:
    print("Morning")
elif hour < 18:
    print("Afternoon")
elif hour < 24:
    print("Night")
else:
    print("Invalid hour")
```

12. Multi-elif: Classify a number by digit count


```python
num =  int(input("Enter number:"))
abs_num = abs(num)
if abs_num < 10:
    print("Single-digit")
elif abs_num < 100:
    print("Two-digit")
else:
    print("Three or more digits")
```

13. Check if a given string is uppercase, lowercase, or mixed


```python
text =  str(input("Enter:"))
if text.isupper():
    print("All uppercase")
elif text.islower():
    print("All lowercase")
else:
    print("Mixed case")
```

14. Multi-elif: Determine shipping cost by weight category

15. Multi-elif: Classify an angle in degrees


```python
angle = int(input("Enter angle:"))
if angle < 90:
    print("Acute")
elif angle == 90:
    print("Right")
elif angle < 180:
    print("Obtuse")
elif angle == 180:
    print("Straight")
else:
    print("Reflex")
```

16. Check if a string is a palindrome


```python
text =  str(input("Enter:"))
if text and text == text[::-1]:
    print("Palindrome")
else:
    print("Not palindrome")
```

18. Check if a number is exactly divisible by 2, 3, or 5

19. Multi-elif: Classify a day number


```python
day =  int(input("Enter day number:"))
if day >= 1 and day <= 5:
    print("Weekday")
elif day == 6:
    print("Saturday")
elif day == 7:
    print("Sunday")
else:
    print("Invalid")
```

19. Nested if: If a character is a digit, check if it's even or odd digit.


```python
ch = '4'
if ch.isdigit():
    digit_value = int(ch)
    if digit_value % 2 == 0:
        print("Even digit")
    else:
        print("Odd digit")
else:
    print("Not a digit")
```
