```python
dict = {1:{'name':"shubham kumar", 'age':22, 'status': "unemployed"},
        2:{'name' :"vikash bhardwaj",'age': 23, 'status': "employed"},
        3:{'name':"sarthak jadoun",'age': 21, 'status': "employed"},
        4:{'name':"shivam kumar", 'age':21, 'status': "self-employed"}} # key value both in quotes
print(dict)
```

```python
# Accesssing
print(dict)
print(dict[2])
print(dict[3])
print(dict[4]['name']) #key bhi quotes m
```

```python
# adding , updating
dict[5]={"earning": 500000}
print(dict)
dict[5]['name'] = 'shashank patel'
print(dict)
dict[5]['status'] = 'employed'
print(dict)
```

```python
# deleting
print(dict)
del dict[5]['earning']
print(dict)
```

```python
# going through the data
print(dict)
print("_"*10)
for i in dict.keys():
  print(i,dict[i]['name'])

print("_"*10)
for i in dict.keys():
  print(i,dict[i]['name'], dict[i]['status'] )


print("_"*10)
for i in dict.items():
  print(i[0])

print("_"*10)
for i in dict.items():
  print(i[1])

print("_"*10)
for i in dict.items():
  print(i[1]["name"])
```
