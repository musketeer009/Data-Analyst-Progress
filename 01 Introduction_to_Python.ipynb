### 3. Operators

#### 3.1) Arithmetic

##### 3.1.1) Int - Int

```python
print(10+8)
print(10-8)
print(10*8)
print(10/8)
print(10//8)
print(10%8)
print(10**8)
```

##### 3.1.2) Int - Float

```python
print(10+8.1)
print(10-8.1)
print(10*8.1)
print(10/8.1)
print(10//8.1)
print(10%8.1)
print(10**8.1)
```

##### 3.1.3) Int - String

```python
print(12*'Shubham')
```

##### 3.1.4) Int - Boolean

```python
print(10 + True)
print(10 - False)
print(10 * True)
print(10 / True)
print(10 // True)
print(10 % True)
print(10 ** True)
```

```python
print(True*'Shubham')
print(False*'Shubham')
print('Shubham'*True)
print('Shubham'*False)
```
