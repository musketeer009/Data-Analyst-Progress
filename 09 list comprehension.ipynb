```python
# square of elements
list=[1,2,3,4,5]
for i in list:
  print(i**2)

list = [i**2 for i in list]
print(list)
list = [i**2 for i in list]
print(list)
```

```python
# findin first 10 even numbers
list1 = [i for i in range(20) if i%2==0]
print(list1)
```

```python
# table of n
n = int(input("enter  no:"))
print([i for i in range(n,n*10+1) if i%n==0])
```

```python
#multi dimension

print([[i for i in range(1,6)] for i in range(3)])
```

```python
# flatten
list = [[i for i in range(1,6)] for i in range(3)]
print(list)
print([i for j in list for i in j])
```

```python
# no is even or odd
list1= ["even" if i%2==0 else "odd" for i in range(1,10)]
print(list1)
```

```python
# pairing two numbers
print([[(x,y) for x in range(1,4)] for y in range(1,3)])
```
