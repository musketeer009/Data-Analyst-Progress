```python
# creating set
my_set = {1,2,3,4,5,6,"shubham"}
print(my_set)
print(type(my_set))
```

```python
# Adding
my_set.add(7)
print(my_set)


my_set.add("kumar")
print(my_set)
my_set.add("kumar")
print(my_set)

my_set.add(1)
my_set.add(1)
print(my_set)
```

```python
# pop
print(my_set)
poped= my_set.pop()
print(poped)
print(my_set)# 1 is


# we can convert set into list and then we can remove by index
```

```python
# removing
print(my_set)
my_set.discard(5)
print(my_set)
```

```python
# iterating

for i in my_set:
  print(i)

print(len(my_set))
```

```python
# checking
print("shubham" in my_set)
```

```python
# set operations
set1 = {1,2,3,7,5,9,7}
set2 = {4,6,0,8,3,6,7,9}
print(set1)
print(set2)
print("union")
print(set1.union(set2))
print(set1 | set2)

print(set1)
```

```python
print("intersect")
print(set1.intersection(set2))
print(set1 & (set2))

```

```python
# difference
print(set1.difference(set2))
print(set1 - (set2))
```

```python
# symm diff
print(set1.symmetric_difference(set2))
print(set1 ^ (set2))
```

```python
# clearing
print(my_set)
my_set.clear()
print(my_set)
```
