```python
# generating square
dict = {i: i**2 for i in range(1,11)}
print(dict)
```

```python
# generating cube
dict = {i: i**3 for i in range(1,11)}
print(dict)
```

```python
# generating square if even
dict = { i: i**3  for i in range(1,11) if i%2==0}
print(dict)
```

```python
list = ["Apple", "Papaya","Cherry","Grappes"]
dict = {i: list[i-1] for i in range(1,len(list)+1)}
print(dict)
```

```python
dict = {list[i]: len(list[i]) for i in range(len(list))}
print(dict)

dict = {i:len(i) for i in list}
print(dict)

#reverse
dict = {len(i): i for i in list}
print(dict)  # update if keys same
```

```python
# generate num_1- num_100
dict = {"num_" + str(i) : i for i in range(10)}
print(dict)


dict = {"num_" + str(i) : i for i in range(100) if i%5==0}
print(dict)

```

```python
# shortlisting
dict = {m:n for m,n in dict.items() if n%7==0}
print(dict)
```

```python
list1 = ['a', 'b', 'c', 'd']
list2 = [1,2,3,4]
dict = {list2[i]: list1[i] for i in range(len(list1))}
print(dict)
```

```python
# nested assign tuple as a key

list = [[1,2,3], [4,5,6],[7,8,9]]
dict = {(i,j): list[i][j] for i in range(len(list[0])) for j in range(len(list[0]))}
print(dict)


# reverse
dict = { list[i][j]: (i,j) for i in range(len(list[0])) for j in range(len(list[0]))}
print(dict)
```
