```python
dict= {1: "Alice",2: "Bob",3: "Charlie",4: "David",5:"Amit", 6:"suresh"}
print(dict[3])
```

```python
# accessing values
dict= {1: "Alice",2: "Bob",3: "Charlie",4: "David",5:"Amit", 6:"suresh", "harsh": 25, True: 22, 5.5:55, False: True}
print(dict[3])
print(dict["harsh"])
print(dict[5.5])
print(dict[True])
print(dict[False])
print(dict[1]) # changed
print(dict[0]) #exception
print(dict.get(4))
```

```python
# adding and updatig and deleting values
dict[7] = "Raghuv"
print(dict)
dict[7] = "Raghvendra"
print(dict)
del dict[7]
print(dict)
```

```python
# clear dictionary
print(dict)
dict.clear()
print(dict)
```

```python
#iterating
dict ={1: 'shubham', 2: 'Bob', 3: 'Charlie', 4: 'David', 5: 'Amit', 6: 'suresh', 'harsh': 25, 5.5: 55, False: True}
print(dict)
print(dict.keys())
print(dict.values())
```

```python
# iterating using for loop
for i in dict.keys():
  print(i, dict[i])
```

```python
# items keyword
print(dict.items())# return ans in list form and consider pair as 1

for i in dict.items():
  print(i[0], i[1])
print("-"*5)

# what!!!!!!!!!!!!!!!
for i,j in dict.items():
  print(i, j)
```

```python
# check if key is present
print(dict)
print(1 in dict)
print("charlie" in dict)
print(11 in dict)
```

```python
# updating dictionary
dict1= {1:"a", 2:"b", 3:"c"}
dict2= {1:"A", 2:"B", 3:"C"}

dict1.update(dict2)
print(dict1)
print(dict2)
```
