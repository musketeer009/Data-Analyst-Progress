```python
l1 =[1,2,3,4,5]
def square(l):
   return [i**2 for i in l]

def cube(l):
   return [i**3 for i in l]

def adding_lists(l1):
  sq= square(l1)
  cu= cube(l1)
  return [sq[i] +cu[i] for i in range(len(l1))]


print(square(l1))
print(cube(l1))
print(adding_lists(l1))
```
