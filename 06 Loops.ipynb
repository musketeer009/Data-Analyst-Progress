```python
# all possibilities of rolling two dice


for k in range(2,13):
  count=0
  for i in range(1,7):
    for j in range(1,7):
      if( i+j==k):

       count+=1

  a = round((count/(6*6))*100,2)
  print(k,"=", str(a) +"%" )
```

```python
# with 3 dice
for k in range(3, 19):
  count=0
  for x in range(1,7):
    for y in range(1,7):
        for z in range(1,7):
          if( x+y+z==k):

           count+=1

  a = round((count/(6*6*6))*100,2)
  print(k,"=", str(a) +"%" )
```
